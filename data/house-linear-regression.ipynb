```python
# IMPORTS
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Reading the dataset
df = pd.read_csv('house.csv')

print(df.info())
print(df.head())
```

```python
# plotting the scatter plot to check the relationship between loyer and surface
sns.lmplot(x='loyer', y='surface', data=df)
```

```python
# Data cleaning
df.fillna(method='ffill', inplace=True)
```

```python
# Separating the data into features and target
X = np.array(df['surface']).reshape(-1, 1)
y = np.array(df['loyer']).reshape(-1, 1)

# Splitting the data into training and test sets 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

regr = LinearRegression()
regr.fit(X_train, y_train)
print(regr.score(X_test, y_test))
```

```python
# Exploring our results
y_pred = regr.predict(X_test)
plt.scatter(X_test, y_test, color='b')
plt.plot(X_test, y_pred, color='k')

plt.show()
```
